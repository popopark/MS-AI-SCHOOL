```python
# 타입의 전환도 무척 간단하다.

a = set((1,2,3))
print(type(a))
b = list(a)
print(type(b))
c = tuple(b)
print(type(c))
d = set(c)
print(type(d))
```

```python
# in 연산자를 통해서 해당 값이 있는지 확인 가능하다.

a = set((1,2,3))
print(1 in a)     # in 연산자는 데이터 타입에 상관없이 사용 할 수 있다.
```

```python
color = {'apple' : 'red', 'banana' : 'yellow'}
print(color)
```

```python
print(color['apple'])

#print(color[0])  # dict 타입은 인덱스 없습니다.

#print(color['cherry'])
```

```python
color['cherry'] = 'red'
print(color)
```

```python
color['apple'] = 'green'
print(color)
```

```python
print(color.keys())
print(color.values())
print(color.items())
```

```python
# for 문을 사용해서 데이터 가져와보기기

for k in color.keys():
  print(k)

for tmp in color.values():
  print(tmp)

for i in color.items():
  print(i)

for (k, v) in color.items():
  print(k, v)
```

```python
# Dict에서 불필요한 아이템의 삭제

del color['cherry']
print(color)
```
