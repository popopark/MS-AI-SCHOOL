```python
# Function

def Times(a, b):
  result = a * b
  return result
```

```python
print(Times)
```

```python
print(Times(10, 20))
```

```python
# 메모리 주소를 확인하는 명령
print(id(Times))
```

```python
# 변수의 범위

g = 10
def sum2(x, y):
  global g 
  g = g+1
  return x, y

print(sum2(x, 20))
print(g)
```

```python
# 함수의 인자 (Parameter)

def Times(a=10, b=20):      # 디폴트 값을 지정
  return a * b

print(Times(100, 200))
print(Times(100))
print(Times())
print(Times(b=200))
```

```python
def connectURL(protocol='https', server='naver.com', port='80'):
  address = protocol + '://' + server + ':' + port
  return address

print(connectURL(port='8080'))
```
